```python
from transformers import BertTokenizer, TFBertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained("bert-base-uncased")

text = "Eat apple"
encoded_input = tokenizer(text, return_tensors='tf')
output = model(encoded_input)
print(output.last_hidden_state.shape) # (1, 4, 768) [CLS]eat apple[SEP]
print("contextualized representation of the word 'apple' in the sentence within the context of the sentence, not just its standalone embedding.")
print(output.last_hidden_state[:, 0])

# Get token ID for "apple"
token_id = tokenizer.convert_tokens_to_ids(["apple"])[0]

# Access embedding table
embedding_weights = model.get_input_embeddings().weight.numpy()

# Get embedding for "apple"
embedding = embedding_weights[token_id]
print("Just tandalone embedding of 'apple'.")
print(embedding)
```
